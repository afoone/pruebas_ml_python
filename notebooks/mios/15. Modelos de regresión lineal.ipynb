# Modelos de regresión lineal

## Modelo con datos simulados

y = a + bx

x será una serie de datos distribuídos con una normal de media 1.5 y desviación estándar 2.5

la y estimada será 

y<sub>e</sub> = 5 + 1.9x + e

siendo e distribuido con una normal N(0.0, 0.8)



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(1.5,2.5, 100)
```

```python
x
```

```python
res = np.random.normal(0,0.8,100)
```

```python
res
```

```python
# Esta es la predicción
y_pred = 5 + 1.9*x
```

```python
y_pred
```

```python
y_real = 5+ 1.9*x +res
```

```python
y_real
```

```python
plt.plot(y_real)
```

```python
plt.plot(y_pred)
```

```python
plt.hist(y_real, bins=40)
```

```python
my_dict = {
        "x":x.tolist(),
        "y":y_real.tolist(),
        "prediccion": y_pred.tolist()
    }
```

```python
my_dict
```

```python
data = pd.DataFrame(
    my_dict
)
```

```python
data.head()
```

```python
y_mean = [np.mean(y_real) for i in range(0, len(y_real))]
```

```python
# grafico para juntar todo en el mismo sitio

%matplotlib inline
plt.plot(x,y_mean, "g")
plt.plot(x,y_pred)
plt.plot(x,y_real, "ro")
plt.title("Valor actual vs Predicción")
```

```python
media_y = np.mean(y_real)
data["SSR"] = (data["prediccion"]-media_y)**2
data["SSD"] = (data["prediccion"]-data["y"])**2
data["SST"] = (data["y"] - media_y)**2
data["media"] = media_y
```

```python
data.head()
```

```python
ssr = sum(data['SSR'])
ssr
```

```python
ssd = sum(data['SSD'])
ssd
```

```python
sst = sum(data['SST'])
sst
```

```python
r2 = ssr / sst
r2
```

```python
# Los errores cuadrados se tienen que distribuir con una normal
# Como es cuadrado... sale media normal en plan cuadrático
plt.hist(data["SSD"])
```
