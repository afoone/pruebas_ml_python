# Distribución uniforme

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.uniform(0, 100, 1000)
```

```python
# según los datos que genero será más o menos evidente
plt.hist(data, bins=10)
```

```python
data = np.random.uniform(0, 100, 1000000)
# según los datos que genero será más o menos evidente
plt.hist(data, bins=10)
```

# Distribución uniforme

```python
data = np.random.normal(5,2, 500)
```

```python
plt.hist(data, bins=20)
```

```python
data = np.random.normal(5,2, 500000)
plt.hist(data, bins=50)
```

```python
#Esto es la normal tipificada (o típica), esto es media cero y desviación típica 1
data = np.random.normal(0,1, 500000)
plt.hist(data, bins=500)
```

```python
# Aquí podemos ver los datos
plt.plot(data[0:100])
```

```python
# podemos pintar los datos ordenados

plt.plot(sorted(data[9:1000]))
```

```python
data = np.random.normal(5,2, 5000)
plt.plot(sorted(data))
```
