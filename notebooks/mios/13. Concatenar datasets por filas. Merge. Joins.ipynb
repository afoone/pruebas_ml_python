### Join Types

```python
# Vamos a eliminar paises y deportes de algunos athletas , así como algunos registros de 'medals'
athletes_to_remove = np.random.choice(medals['Athlete'], 10).tolist()
athletes_to_remove.append("Michael Phelps")
athletes_to_remove
```

```python

medals = medals[~medals['Athlete'].isin(athletes_to_remove)]
```

```python
# Vamos a
countries_to_remove = np.random.choice(countries['Country'], 10).tolist()
countries_to_remove
```

```python
countries = countries[~countries['Country'].isin(countries_to_remove)]
```

```python
merged_inner = pd.merge(left=countries, right=medals, left_on="Athlete", right_on="Athlete", how="inner")
```

```python
len (merged_inner)
```

```python
merged_left = pd.merge(left=countries, right=medals, left_on="Athlete", right_on="Athlete", how='left')
```

```python
len (merged_left)
```

```python
merged_left.head()
```

```python
merged_right=pd.merge(left=countries, right=medals, left_on="Athlete", right_on="Athlete", how="right")
```

```python
len(merged_right)
```

```python
merged_right.head()
```

```python
merged_right[merged_right['Country'].isna()]
```

```python
merged_left[merged_left['Gold Medals'].isna() ]
```

```python
merged_full = pd.merge(left=countries, right=medals, left_on='Athlete', right_on='Athlete', how='outer')
```

```python
merged_full[merged_full['Gold Medals'].isna()]
```
